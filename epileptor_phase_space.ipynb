Unstable fixed point

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


I = 3.1
x0 = -1.8
tau0 = 10.0

x = np.r_[-3:1:0.01]
z = np.r_[2:5:0.01]
xx, zz = np.meshgrid(x,z)
dx = np.zeros([len(z),len(x)])
dz = np.zeros([len(z),len(x)])

for i in range(len(z)):
    for j in range(len(x)):
        dx[i,j] = 1 - x[j]**3 - 2*x[j]**2 - z[i] + I
        dz[i,j] = 1/tau0*(4*(x[j] - x0) - z[i])

plt.figure(figsize=(10,8))
# plt.quiver(xx,zz,dx,dz,scale_units='xy',minlength=1e-5)
plt.streamplot(x,z,dx,dz,density=3.0)

x = np.r_[-3:1:0.01]
xNC = 1 - x**3 - 2*x**2 + I
zNC = 4*(x-x0)
plt.plot(x,xNC)
plt.plot(x,zNC)
plt.xlabel('x',fontsize=15.0)
plt.ylabel('z',fontsize=15.0)
plt.xlim([-3,1])
plt.ylim([2,5])
plt.legend(['$\dot x=0$','$\dot z=0$'],loc=1,fontsize=15.0)
plt.show(block=False)
```
